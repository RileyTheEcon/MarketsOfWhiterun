## Preference Parameter distributions

  

```python
####    EXPERIMENT 2    ####
# Experiment 2 A
sessionLog.print('Experiment 2.A (Log-Normal) start -- allow ~5 min runtime')

m = 0
v = 1

exp2a = e2.experiment2(lambda : np.random.lognormal(mean=m, sigma=v, size=4), verbose=False) # allow ~2 min runtime
exp2a.runSimulation()
sessionLog.print('\n')
```

```python
# Experiment 2 B
sessionLog.print('Experiment 2.B (Gamma) start -- allow ~5 min runtime')

alpha = 1
beta  = 1

exp2b = e2.experiment2(lambda : np.random.gamma(alpha,1/beta, size=4), verbose=False) # allow ~2 min runtime
exp2b.runSimulation()
sessionLog.print('\n')
```

```python
# Experiment 2 C
sessionLog.print('Experiment 2.C (Dirichlet) start -- allow ~5 min runtime')

exp2c = e2.experiment2(lambda : np.random.dirichlet([1.0,1.0,1.0,1.0]), verbose=False) # allow ~2 min runtime
exp2c.runSimulation()
sessionLog.print('\n')

####
```

```python
# Experiment 2 D
sessionLog.print('Experiment 2.D (Dirichlet-adjusted) start -- allow ~5 min runtime')

vecParam = [0.285580,  0.157223,  0.495946,  0.061252]
exp2d = e2.experiment2(
    lambda : np.random.dirichlet(vecParam),
    verbose=False
    )
exp2d.runSimulation()
sessionLog.print('\n')
    
```
